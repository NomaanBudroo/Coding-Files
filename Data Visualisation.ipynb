```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset without column names
file_path = "DMV302_Assessment_3_Dataset2.csv"
data = pd.read_csv(file_path, header=None)

# Convert the first column (Date) to datetime format
data[0] = pd.to_datetime(data[0], dayfirst=True)

# Visualize temperature variations over the financial year using a line plot
plt.figure(figsize=(12, 6))
sns.lineplot(x=0, y=1, data=data, marker='o', label='Actual Avg Temp')
sns.lineplot(x=0, y=2, data=data, marker='o', label='Actual Min Temp')
sns.lineplot(x=0, y=3, data=data, marker='o', label='Actual Max Temp')
plt.title('Temperature Variations Over the Financial Year')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

# Visualize the distribution of rainfall using a histogram
plt.figure(figsize=(10, 6))
sns.histplot(data[8], bins=20, kde=True)
plt.title('Distribution of Actual Rainfall')
plt.xlabel('Rainfall (mm)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Box plot for temperature
plt.figure(figsize=(10, 6))
sns.boxplot(data=data.iloc[:, 1:4], palette='viridis')
plt.title('Box Plot of Temperature Data')
plt.ylabel('Temperature (°C)')
plt.xticks(ticks=[0, 1, 2], labels=['Actual Avg Temp', 'Actual Min Temp', 'Actual Max Temp'])
plt.tight_layout()
plt.show()
```

```python
# Scatter plot for temperature vs. rainfall
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data[1], y=data[8], label='Actual Avg Temp')
sns.scatterplot(x=data[2], y=data[8], label='Actual Min Temp')
sns.scatterplot(x=data[3], y=data[8], label='Actual Max Temp')
plt.title('Scatter Plot of Temperature vs. Rainfall')
plt.xlabel('Temperature (°C)')
plt.ylabel('Rainfall (mm)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Monthly average temperature
data['Month'] = data[0].dt.month
monthly_avg_temp = data.groupby('Month')[1].mean()

plt.figure(figsize=(10, 6))
sns.barplot(x=monthly_avg_temp.index, y=monthly_avg_temp.values, palette='coolwarm')
plt.title('Monthly Average Temperature')
plt.xlabel('Month')
plt.ylabel('Average Temperature (°C)')
plt.tight_layout()
plt.show()
```

```python
# Rainfall distribution
plt.figure(figsize=(10, 6))
sns.histplot(data[9], kde=True, color='skyblue')
plt.title('Distribution of Daily Rainfall')
plt.xlabel('Rainfall (mm)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Rainfall vs. Temperature correlation
plt.figure(figsize=(10, 6))
sns.heatmap(data.iloc[:, [1, 9]].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Rainfall vs. Temperature Correlation')
plt.tight_layout()
plt.show()
```
